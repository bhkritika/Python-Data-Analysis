## Charts

### Gender Distribution 

```python
sns.countplot(data = data, x = "Gender")
plt.show()
```

```python
plt.figure(figsize=(6,5))
axes = sns.countplot(data = data, x = "Gender")
axes.bar_label(axes.containers[0])
plt.title("Distribution of Gender")
plt.show()
```

From chart we analysed that: 
The number of females in the dataset are more than the number of males present in the dataset. 

```python
# Distribution of Math Scores
plt.figure(figsize=(10, 6))
sns.histplot(data['MathScore'], bins=20, kde=True)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of Scores by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='MathScore', data=data)
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

```python
# Pairplot of Scores
sns.pairplot(data[['MathScore', 'ReadingScore', 'WritingScore']])
plt.show()
```

### Correlation Matrix 

```python
# Correlation matrix
corr_matrix = data[['MathScore', 'ReadingScore', 'WritingScore']].corr()
print(corr_matrix)

# Create a heatmap with proper formatting
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5, annot_kws={"size": 10})
plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()  # Adjust layout to ensure everything fits well
plt.show()
```

### GroupBy 

```python
grp = data.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
```

```python
print(grp)
```

### HeatMap 

```python
plt.figure(figsize=(6,5))
sns.heatmap(grp, annot = True)
plt.show()
```

```python
# Define colors for the bars
colors = ['steelblue', 'mediumseagreen', 'salmon']

# Visualize the average scores by Parent Education
grp.plot(kind='bar', figsize=(12, 6), color = colors)
plt.title('Average Scores by Parent Education')
plt.xlabel('Parent Education')
plt.ylabel('Average Score')
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(20, 8)) 

# WritingScore boxplot
plt.subplot(1, 3, 3)
sns.boxplot(x='EthnicGroup', y='WritingScore', data=data, palette='Set3')
plt.title('Writing Score by Ethnic Group', fontsize=16)
plt.xlabel('Ethnic Group', fontsize=14)
plt.ylabel('Writing Score', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
```

### Distribution Of Ethnic Groups 

```python
groupA = data.loc[(data['EthnicGroup'] == "group A")].count()
print(groupA)
```

```python
groupB = data.loc[(data['EthnicGroup'] == "group B")].count()
groupC = data.loc[(data['EthnicGroup'] == "group C")].count()
groupD = data.loc[(data['EthnicGroup'] == "group D")].count()
groupE = data.loc[(data['EthnicGroup'] == "group E")].count()
```

```python
mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mylist)
```

```python
order = ["group A", "group B", "group C", "group D", "group E"]
mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mylist, labels = order, autopct= "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
